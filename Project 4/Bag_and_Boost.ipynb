```python
# 1 = male, 0 = female
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
file = "customer_shopping_data.csv"
shopping = pd.read_csv(file)
```

```python
shopping.head(5)
```

```python
shopping.drop(columns = "Unnamed: 9", inplace = True)
```

```python
shopping.head(25)
```

```python
shopping.invoice_date = pd.to_datetime(shopping.invoice_date)
shopping['Day of week'] = shopping.invoice_date.dt.dayofweek
shopping.head(25)
```

```python
shopping.dtypes
```

```python
shopping.drop(columns = ['invoice_no','customer_id'], inplace = True)
```

```python
shopping.head()
```

```python
shopping['Day of week'].replace(to_replace = 6, value = 'Sunday', inplace =True)
shopping['Day of week'].replace(to_replace = 5, value = 'Saturday', inplace = True)
shopping['Day of week'].replace(to_replace = 4, value = 'Friday', inplace = True)
shopping['Day of week'].replace(to_replace = 3, value = 'Thursday', inplace = True)
shopping['Day of week'].replace(to_replace = 2, value = 'Wednesday', inplace = True)
shopping['Day of week'].replace(to_replace = 1, value = 'Tuesday', inplace = True)
shopping['Day of week'].replace(to_replace = 0, value = 'Monday', inplace = True)
shopping
```

```python
stuff = shopping["Day of week"].unique()
stuff2 = shopping["category"].unique()
stuff3 = shopping['payment_method'].unique()
stuff4 = shopping['shopping_mall'].unique()
stuff4
```

```python
one_hot_encoded = pd.get_dummies(shopping["Day of week"], columns = ['Sunday', 'Saturday', 'Tuesday', 'Wednesday', 'Friday', 'Thursday',
       'Monday'])
one_hot_encoded
```

```python
shopping.drop(columns = ['invoice_date'], inplace = True)
```

```python
shopping
```

```python
shopping['gender'].replace(to_replace = 'Male', value = 1 , inplace = True)
shopping['gender'].replace(to_replace = 'Female', value = 0, inplace = True)
```

```python
new_shop = pd.get_dummies(shopping)
```

```python
new_shop.columns
```

```python
new_shop
```

```python
X = new_shop.drop(columns = ['gender'])
y = new_shop['gender']
```

```python
from sklearn.model_selection import train_test_split
target_names = ["Female", "Male"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=33)
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(random_state=1).fit(X_train_scaled, y_train)
y_pred = clf.predict(X_test_scaled)
print(classification_report(y_test, y_pred, target_names=target_names))
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
clf = ExtraTreesClassifier(random_state=36).fit(X_train_scaled, y_train)
y_pred = clf.predict(X_test_scaled)
print(classification_report(y_test, y_pred, target_names=target_names))
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
clf = AdaBoostClassifier(random_state=1).fit(X_train_scaled, y_train)
y_pred = clf.predict(X_test_scaled)
print(classification_report(y_test, y_pred, target_names=target_names))
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```
